```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Dataset from your image
data = {
    'Year': ['2020', '2021', '2022', '2023', '2024'],
    'Revenue ($M)': [523964, 559151, 572754, 611289, 648125],
    'Employee Count': [2200000, 2300000, 2300000, 2100000, 2100000],
    'Operating Income ($M)': [20568, 22548, 25942, 20428, 27012],
    'Gross Profit ($M)': [129359, 138836, 143754, 147568, 157983],
    'SG&A Expenses ($M)': [108791, 116288, 117812, 127140, 130971],
    'Current Ratio': [0.84, 0.95, 0.87, 0.85, 0.85]
}

df = pd.DataFrame(data)
df.set_index('Year', inplace=True)

# Step 2: Normalize for heatmap scaling (optional)
df_scaled = df.copy()
for col in df.columns:
    df_scaled[col] = (df[col] - df[col].min()) / (df[col].max() - df[col].min())

# Step 3: Heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df_scaled.T, annot=df.T, fmt=".0f", cmap='Blues', linewidths=0.5, cbar=False)

plt.title("üî• Walmart Strategic Metric Heatmap (2020‚Äì2024)", fontsize=14)
plt.ylabel("Key Metrics")
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
